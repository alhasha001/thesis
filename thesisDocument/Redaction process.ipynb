# The process of Redaction / Whitewashing 


## Within the Dutch government


First of all, what is data redaction or whitewashing? It is the process of searching for private sensitive or confidential information and hiding it. What the government usually must hide from published government papers are details such as people's names, social security numbers, Id number, driver's license number, banking, and financial information. The Dutch civil servants use the Wob-help tooling which refers to the software used while handling the Wob requests. 

The Wob-help tooling software consists of two parts. The first part is a search engine that goes through the chosen documents using a text analysis algorithm to recognize the private sensitive information. The second part is a utility software that is used for tasks such as redacting sensitive information or / and deduplicating information. The Wob civil servant eventually decides which information to be disclosed or hidden.
